```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ = pd.read_csv(r"C:\Users\Laptop\Documents\watson_healthcare_modified.csv")
df_
```

```python
df_["MonthlyIncome"].to_frame()
```

```python
#Average Salary For Each Department
df_av_sal_dept = df_.groupby("Department")["MonthlyIncome"].mean().to_frame().sort_values(by="MonthlyIncome", ascending= False)
df_av_sal_dept
```

```python
df_av_sal_dept.plot(kind ="bar", xlabel ='Department' , ylabel='Avg_Income', title = "Average Salary Income By Department")
```

```python
# Employees with more than 10 years of stay
df_['YearsAtCompany'].to_frame()
```

```python
more_than_10_years= df_["EmployeeID"][df_["YearsAtCompany"]> 10]
more_than_10_years.to_frame()
```

```python
# Department and their average perfromance ratings
df_avg_rating_by_dept = df_.groupby("Department")["PerformanceRating"].mean().to_frame().sort_values(by="PerformanceRating",ascending = False)
df_avg_rating_by_dept
```

```python
#Number Of Employees In Each Department
df_count_empl = df_.groupby("Department")["EmployeeID"].count().to_frame().sort_values(by= "EmployeeID", ascending = False)
df_count_empl
```

```python
df_count_empl.plot(kind ="bar", xlabel ='Department' , ylabel='Employee ID', title = "Number Of Employees In Each Department")
```

```python
#Correlating Years at company with percentage salary hike
df_YAT_PSH = df_[["YearsAtCompany","PercentSalaryHike"]]
plt.figure(figsize = (10,8))
sns.heatmap(df_YAT_PSH.corr(),annot =True,cmap = "Blues")
plt.title("Years at company with percentage salary hike")
plt.show()
```

```python
#Correlating Years at company with Monthly Income
df_YAT_MI = df_[["YearsAtCompany","MonthlyIncome"]]
plt.figure(figsize = (10,8))
sns.heatmap(df_YAT_MI.corr(),annot =True,cmap = "Blues")
plt.title("Years at company with Monthly Income")
plt.show()
```

```python
#Correlating Monthly Rate with Monthly Income
df_MI_MR = df_[["MonthlyRate","MonthlyIncome"]]
plt.figure(figsize = (10,8))
sns.heatmap(df_MI_MR.corr(),annot =True,cmap = "Blues")
plt.title("Monthly Rate with Monthly Income")
plt.show()
```

```python
#Correlating Distance From Home with Performance Rating
df_DFH_PR = df_[["DistanceFromHome","PerformanceRating"]]
plt.figure(figsize = (10,8))
sns.heatmap(df_DFH_PR.corr(),annot =True,cmap = "Blues")
plt.title("Distance From Home with Performance Rating")
plt.show()
```
